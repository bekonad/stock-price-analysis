```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("✅ Libraries imported successfully.")
```

```python
df = pd.read_csv("../data/raw/stock_prices.csv")

print("✅ Dataset loaded successfully.")
```

```python
print("First 5 Rows:")
print(df.head())
```

```python
print("Dataset Shape:")
print(df.shape)

print(f"\nRows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
print("="*50)
print("Dataset Information")
print("="*50)

df.info()
```

```python
print("="*50)
print("Missing Values")
print("="*50)

print(df.isnull().sum())
```

```python
df_clean = df.copy()

print("✅ Working copy created.")
```

```python
df_clean["date"] = pd.to_datetime(df_clean["date"])

print("✅ Date column converted.")

print("\nData Types:")
print(df_clean.dtypes)
```

```python
df_clean = df_clean.sort_values(
    by=["symbol", "date"]
)

print("✅ Dataset sorted by symbol and date.")

print("\nFirst 5 Rows:")
print(df_clean.head())
```

```python
df_clean[
    ["open", "high", "low"]
] = (
    df_clean.groupby("symbol")
    [["open", "high", "low"]]
    .ffill()
)

print("✅ Forward fill completed.")
```

```python
print("Missing Values After Forward Fill:")
print(df_clean.isnull().sum())
```

```python
print("Rows with Remaining Missing Values:")

print(
    df_clean[
        df_clean.isnull().any(axis=1)
    ]
)
```

```python
df_clean = df_clean.dropna()

print("✅ Remaining missing rows removed.")
```

```python
print("Missing Values After Final Cleaning:")

print(df_clean.isnull().sum())
```

```python
duplicates = df_clean.duplicated().sum()

print(f"Duplicate Rows: {duplicates}")
```

```python
df_clean = df_clean.drop_duplicates()

print("✅ Duplicate rows removed (if any existed).")
```

```python
print(
    f"Number of Companies: "
    f"{df_clean['symbol'].nunique()}"
)
```

```python
print("First 10 Companies:")

print(
    df_clean["symbol"]
    .unique()[:10]
)
```

```python
print("="*50)
print("Summary Statistics")
print("="*50)

print(df_clean.describe())
```

```python
df_clean.to_csv(
    "../data/processed/stock_prices_cleaned.csv",
    index=False
)

print("✅ Cleaned dataset saved successfully.")
```

```python
print("="*60)
print("DATA CLEANING COMPLETED")
print("="*60)

print(f"Rows: {df_clean.shape[0]}")
print(f"Columns: {df_clean.shape[1]}")
print(f"Companies: {df_clean['symbol'].nunique()}")

print("\nMissing Values:")
print(df_clean.isnull().sum())

print("\nDuplicate Rows:")
print(df_clean.duplicated().sum())

print("\nData Types:")
print(df_clean.dtypes)

print("="*60)
print("Dataset is ready for Exploratory Data Analysis (EDA)")
print("="*60)
```
